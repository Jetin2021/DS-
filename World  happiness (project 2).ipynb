```python
#plotting every column with Happiness Rank

features_columns = ['Happiness Score','Standard Error', 'Family', 'Freedom', 'Generosity', 'Dystopia Residual']
features = df[features_columns]
target = df['Happiness Rank']

# plot settings
n_cols = 3
n_rows = len(features_columns)//n_cols + 1
figsize = (14,10)
fig,ax = plt.subplots(n_rows,n_cols,figsize=figsize)
fig.suptitle('Happiness in 2018',fontsize=25)
index = 0

for i in range(n_rows):
    ax[i][0].set_ylabel('Happiness Rank')
    for j in range(n_cols):
        x = features[features_columns[index]]
        ax[i][j].scatter(x,target)
        ax[i][j].set_xlabel(features_columns[index])
        
        # adding line to the data
        fit = np.polyfit(x,target,1)
        ax[i][j].plot(x,fit[0]*x + fit[1],color='red')
        
        index+=1
        if index == len(features_columns):
            break
plt.tight_layout()
plt.show()
```
